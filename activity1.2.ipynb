## Steps

### 1. Frequency Distribution Table

We create a frequency distribution table with 5 equal classes.

| Class Interval  | Frequency |
|-----------------|------------|
| 14.77 - 15.03   | 3          |
| 15.04 - 15.30   | 4          |
| 15.31 - 15.57   | 0          |
| 15.58 - 15.84   | 2          |
| 15.85 - 16.11   | 7          |

```python
import matplotlib.pyplot as plt
# Data for cumulative frequency
class_intervals = ['14.77-15.03', '15.04-15.30', '15.31-15.57', '15.58-15.84', '15.85-16.11']
cumulative_frequencies = [3, 7, 7, 9, 16]
```

### 2. Ogive (Cumulative Frequency) Curve

To construct the ogive, we calculate the cumulative frequencies for each class interval.

| Class Interval  | Frequency | Cumulative Frequency |
|-----------------|------------|----------------------|
| 14.77 - 15.03   | 3          | 3                    |
| 15.04 - 15.30   | 4          | 7                    |
| 15.31 - 15.57   | 0          | 7                    |
| 15.58 - 15.84   | 2          | 9                    |
| 15.85 - 16.11   | 7          | 16                   |


```python
# Plotting the Ogive
plt.figure(figsize=(10, 6))
plt.plot(class_intervals, cumulative_frequencies, marker='o', linestyle='-')
plt.title('Ogive (Cumulative Frequency) Curve')
plt.xlabel('Class Intervals')
plt.ylabel('Cumulative Frequency')
plt.grid(True)
plt.show()
```
